```python
import pandas as pd
import numpy as np
```

```python
#Creating Series of lists to use them as columns for the dataframes

id = pd.Series(['A001','A002','A003','A004','A005'])
project = pd.Series(
    ['Project 1', 'Project 2', 'Project 3', 'Project 4', 'Project 5','Project 6', 'Project 7', 
     'Project 8', 'Project 9', 'Project 10', 'Project 11', 'Project 12', 'Project 13', 'Project 14']
     )
name = pd.Series(["John Alter", "Alice Luxumberg", "Tom Sabestine", "Nina Adgra", "Amy Johny"])
gender = pd.Series(["M","F","M","F","F"])
city = pd.Series(["Paris", "London", "Berlin", "New York", "Madrid"])
age = pd.Series([25,27,29,31,30])
status = ["Finished", "Ongoing", "Failed"]
designation_level = [2,2,3,2,3]
status_list = pd.Series(["Finished", "Ongoing","Finished", "Ongoing", "Finished", "Failed", "Finished", "Failed", "Ongoing", "Finished", "Failed", "Ongoing", "Finished", "Finished" ])
```

```python
# Creating Project Dataframe

project_df = pd.DataFrame({
    "ID" : pd.Series(['A001','A002','A003','A004','A005', 'A002', 'A005', 'A003', 'A001', 'A003', 'A001','A004', 'A004', 'A005']),
    "Project" : project,
    "Cost" : pd.Series([1002000, 2000000, 4500000, 5500000,'' , 680000, 400000, 350000,'' , 300000, 2000000, 1000000, 3000000, 200000]),
    "Status" : pd.Series(["Finished", "Ongoing","Finished", "Ongoing", "Finished", "Failed", "Finished", "Failed", "Ongoing", "Finished", "Failed", "Ongoing", "Finished", "Finished" ])
})
project_df
```

```python
emp_df = pd.DataFrame({
    "ID" : id,
    "Name" : name,
    "Gender" : gender,
    "City" : city,
    "Age" : age,
})
emp_df
```

```python
seni_df = pd.DataFrame({
    "ID" : id,
    "Designation Level" : designation_level,
})
seni_df
```

```python
# Task 1
# making three dataframe in python and save them as .csv files
# Save DataFrames to CSV files
project_df.to_csv('project_data.csv', index=False)
emp_df.to_csv('employee_data.csv', index=False)
seni_df.to_csv('seniority_data.csv', index=False)

print("CSV files created successfully.")
```

```python
#Task 2

# Create the "Project" dataframe
# Assuming 'project_df' from previous cell is the intended data
df = pd.DataFrame(project_df) 

# Convert 'Cost' column to numeric, coercing errors to NaN
df['Cost'] = pd.to_numeric(df['Cost'], errors='coerce')

# Calculate the running average and replace missing values
running_sum = 0
count = 0

for i in range(len(df)):
    if pd.notna(df.loc[i, 'Cost']): # Use 'Cost' as column name is case-sensitive
        running_sum += df.loc[i, 'Cost']
        count += 1
    else:
        if count > 0:
            df.loc[i, 'Cost'] = running_sum / count

print("DataFrame with replaced missing values:")
print(df)
```

```python
# Task 3
# Split First and Last name
# Create the "Employee" dataframe
df = pd.DataFrame(emp_df)

# Split the "Name" column into "First Name" and "LastName"
df[['First Name', 'LastName']] = df['Name'].str.split(' ', n=1, expand=True)  # Use n=1 to split on the first space

# Remove the original "name" column
df.drop('Name', axis=1, inplace=True)

print("DataFrame with split Name columns:")
print(df)
```

```python
# Task 4
# Join all three dataframes in one single dataframe. Name it “Final”
df1 = pd.DataFrame(project_df)
df2 = pd.DataFrame(df)
df3 = pd.DataFrame(seni_df)

# Merge the dataframes on the 'ID' column
Final = df1.merge(df2, on='ID', how='outer').merge(df3, on='ID', how='outer')

print("Final DataFrame with merged data:")
print(Final)
```

```python
# Task 5 Add a new bonus column in the Final dataframe. Give a 5% bonus concerning project cost only to employees who have finished the projects
# #Using 'for' loop to creating a list of Bonus cost
#Later, inserted the list into the final dataframe

results_list=[]

for i, r in Final.iterrows():
    # Convert 'Cost' to numeric, handling potential errors
    cost = pd.to_numeric(r['Cost'], errors='coerce') 
    if(Final['Status'].iat[i] == 'Finished'):
        result = cost + cost * 0.05 if pd.notna(cost) else None # Handle potential NaN values
        results_list.append(result)
    else:
        result = cost if pd.notna(cost) else None # Handle potential NaN values
        results_list.append(result)

Final.insert(9, "Bonus 5%", results_list)
```

```python
# printing Final after adding 5% bonus
print(Final)
```

```python
# Task 6 
#Demote the designation level by 1, whose projects have status “fail”. Delete the employees record whose designation level is above 4.

new_des = []

for i, r in Final.iterrows():
    if(Final["Status"].iat[i] == 'Failed'):
        result=(r['Designation Level']) + 1
        new_des.append(result)
    else:
        result=(r['Designation Level'])-1
        new_des.append(result)
```

```python
# Task 8
# Promote designation level by 1 for the employees whose age is more than 29 years using IF condition.

new_des = []

for i, r in Final.iterrows():
    if(Final["Age"].iat[i] > 29):
        result=(r['Designation Level']) + 1
        new_des.append(result)
    else:
        result=(r['Designation Level'])-1
        new_des.append(result)

Final.insert(6, 'Designation', new_des)
Final.drop('Designation Level', inplace = True, axis = 1)
Final.rename(columns={"Designation":"Designation Level"}, inplace = True)
Final
```
